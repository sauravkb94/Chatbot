The NLTK Vader sentiment analyzer is a rule-based model for sentiment analysis of social media text 1. It is a part of the Natural Language Toolkit (NLTK), which is a popular Python library for working with human language data 2. The Vader model uses a lexicon of words and their sentiment scores to determine the sentiment of a given text 1. It also takes into account punctuation, capitalization, and intensifiers to provide a more accurate sentiment score 1. The Vader model is particularly useful for analyzing short texts such as tweets, reviews, and comments

A **word cloud** is a visual representation of the most frequently used words in a text. The words are arranged in a way that the more frequently a word appears in the text, the larger and more prominent it appears in the word cloud ¹. Word clouds are often used to help identify the most important themes or topics in a text, and can be useful for summarizing large amounts of information in a single image ². They are also used in data visualization to help communicate the results of text analysis ². To create a word cloud, you can use a variety of tools, such as **Wordclouds.com**, which is a free online word cloud generator and tag cloud creator .


```python
def wordcloud(df, label):

    subset=df[df[label]==1]
    text=df.Query.values
    wc= WordCloud(background_color="black",max_words=1000)

    wc.generate(" ".join(text))

    plt.figure(figsize=(20,20))
    plt.subplot(221)
    plt.axis("off")
    plt.title("Words frequented in {}".format(label), fontsize=20)
    plt.imshow(wc.recolor(colormap= 'gist_earth' , random_state=244), alpha=0.98)
# visualising wordcloud
```

**Text-Normalization**

Text normalization is porcess where we change our text and convert the text into a usable form, where we remove all the specail chars and we convert the text in lowercase, also remove any non ASCII characters and alphanumeric values as well. help us to covert in upper to lower,number into words, remove white space, expand abbreviation and also we can remove words(most fequently words).

```python
# Removing special characters
!pip install re
# importing regular expressions
```

```python
import re
```

```python
punc_lower = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x.lower())
# Lower case conversion
```

```python
  remove_n = lambda x: re.sub("\n", " ", x)
# removing \n and replacing them with empty value
```

```python
remove_non_ascii = lambda x: re.sub(r'[^\x00-\x7f]',r' ', x)
# removing non ascii characters
```

```python
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x)
# removing alpha numeric values
```

```python
df['Query'] = df['Query'].map(alphanumeric).map(punc_lower).map(remove_n).map(remove_non_ascii)
# using map function and applying the function on query column
```

```python
df['Response'] = df['Response'].map(alphanumeric).map(punc_lower).map(remove_n).map(remove_non_ascii)
# using map function and applying the function on response column
```

```python
query = 'who are you ?'
def chatbot(query):
    # step:-1 clean
    query = nu.lemmatization_sentence(query)
    # step:-2 word embedding - transform
    query_vector = tfidf.transform([query]).toarray()
    # step-3: cosine similarity
    similar_score = 1 -cosine_distances(factors,query_vector)
    index = similar_score.argmax() # take max index position
    # searching or matching question
    matching_question = df.loc[index]['Query']
    response = df.loc[index]['Response']
    pos_score = df.loc[index]['pos']
    neg_score = df.loc[index]['neg']
    neu_score = df.loc[index]['neu']
    confidence = similar_score[index][0]
    chat_dict = {'match':matching_question,
                'response':response,
                'score':confidence,
                'pos':pos_score,
                'neg':neg_score,
                'neu':neu_score}
    return chat_dict
```
